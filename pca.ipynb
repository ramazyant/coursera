```
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```
initial_df = pd.read_csv("/content/close_prices.csv")
initial_df.head()
```

```
df = initial_df.iloc[:, 1:]
# df.head()
```

```
model = PCA(n_components=10)
model.fit(df)
```

```
total_var = 0
for i, var in enumerate(model.explained_variance_ratio_):
  total_var += var
  if total_var > 0.9:
    break

print(i + 1)
```

```
X_ = pd.DataFrame(model.transform(df))[0]
X_.head()
```

```
df_ = pd.read_csv("/content/djia_index.csv")
df_.head()
# print(df_.iloc[:, 0], df_.iloc[:, 1])
```

```
corr = np.corrcoef(X_, y=df_["^DJI"])
print(f"{corr[1, 0]:.2f}")
```
