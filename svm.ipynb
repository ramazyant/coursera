```
import pandas as pd
import numpy as np
from sklearn.svm import SVC
```

```
df = pd.read_csv("/content/svm-data.csv", header = None)
```

```
X = df.loc[:, 1:]
y = df.loc[:, 0]
```

```
clf = SVC(C=100000, kernel='linear', random_state=241)
```

```
clf.fit(X, y)
```

```
print(" ".join([str(ind + 1) for ind in np.sort(clf.support_)]))
```
