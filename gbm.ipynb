```
X = df.iloc[:, 1:].values
y = df["Activity"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=241)
```

```
def sigmoid(y_pred: np.array) -> List[float]:
  return 1.0 / (1.0 + np.exp(-y_pred))
```

```
def log_loss_res(model, X: np.array, y: np.array) -> List[float]:
  return [log_loss(y, sigmoid(y_pred)) for y_pred in model.staged_decision_function(X)]
```

```
def plot_loss(learning_rate: float, test_loss: List[float], train_loss: List[float]):
  plt.figure()
  plt.plot(test_loss, "g", linewidth=2)
  plt.plot(train_loss, "r", linewidth=2)
  plt.legend(["test", "train"])
  plt.show()
```

```
min_loss_results = {}
for rate in [1, 0.5, 0.3, 0.2, 0.1]:
  model = GradientBoostingClassifier(learning_rate=rate, n_estimators=250, verbose=True, random_state=241)
  model.fit(X_train, y_train)

  train_loss = log_loss_res(model, X_train, y_train)
  test_loss = log_loss_res(model, X_test, y_test)
  plot_loss(rate, test_loss, train_loss)

  min_loss = min(test_loss)
  min_loss_ind = test_loss.index(min_loss) + 1
  min_loss_results[rate] = min_loss, min_loss_ind
```

```
print("overfitting")
```

```
min_loss, min_loss_ind = min_loss_results[0.2]
print(f"{min_loss:.2f} {min_loss_ind}")
```

```
model = RandomForestClassifier(n_estimators=min_loss_ind, random_state=241)
model.fit(X_train, y_train)

y_pred = model.predict_proba(X_test)[:, 1]
test_loss = log_loss(y_test, y_pred)

print(f"{test_loss:.2f}")
```
