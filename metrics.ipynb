```
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve
```

```
df = pd.read_csv("/content/classification.csv")
df.head()
```

```
TP = df[(df['pred'] == 1) & (df['true'] == 1)]
FP = df[(df['pred'] == 1) & (df['true'] == 0)]
FN = df[(df['pred'] == 0) & (df['true'] == 1)]
TN = df[(df['pred'] == 0) & (df['true'] == 0)]
print(len(TP), len(FP), len(FN), len(TN))
```

```
acc = accuracy_score(df['true'], df['pred'])
prec = precision_score(df['true'], df['pred'])
rec = recall_score(df['true'], df['pred'])
f1 = f1_score(df['true'], df['pred'])
print(f"{acc:.2f} {prec:.2f} {rec:.2f} {f1:.2f}")
```

```
df = pd.read_csv("/content/scores.csv")
df.head()
```

```
# logreg = roc_auc_score(df['true'], df['score_logreg'])
# svm = roc_auc_score(df['true'], df['score_svm'])
# knn = roc_auc_score(df['true'], df['score_knn'])
# tree = roc_auc_score(df['true'], df['score_tree'])
# print(f"{logreg:.3f} {svm:.3f} {knn:.3f} {tree:.3f}")
clf_names = df.columns[1:]
scores = pd.Series([roc_auc_score(df['true'], df[clf]) for clf in clf_names], index=clf_names)

print(scores.sort_values(ascending=False).index[0])
```

```
pr_scores = []
for clf in clf_names:
  pr_curve = precision_recall_curve(df['true'], df[clf])
  pr_scores.append(pr_curve[0][pr_curve[1] >= 0.7].max())

print(clf_names[np.argmax(pr_scores)])
```
